```python
import os
import glob
import numpy as np
import shutil
from cellpose import models
from tqdm import tqdm
from skimage.io import imread, imsave

base_input_dir = '/kaggle/input/her2-sish'
base_output_dir = '/kaggle/working/masks'
subfolders = [
    'Amplified samples',
    'Amplified test cases',
    'Non Amplified samples',
    'Non Amplified test cases',
    'Nuclei Annotations'
]

model = models.CellposeModel(gpu=True, pretrained_model='/kaggle/input/cellpose-hil-model/tensorflow2/default/1/cellpose_HITL')
channels = [[0, 3]]
all_done = []

print("Starting mask creation...")

for sub in subfolders:
    input_dir = os.path.join(base_input_dir, sub)
    output_dir = os.path.join(base_output_dir, sub) 

    files = glob.glob(os.path.join(input_dir, '*.png'))
    if not files:
        print(f"No PNG files found in {input_dir}, skipping...")
        continue

    imgs = [imread(f) for f in files]
    all_masks = []

    print(f"\nProcessing {len(imgs)} images from: {sub}")
    for img in tqdm(imgs, desc=f"Creating masks for {sub}", unit="image"):
        masks, _, _ = model.eval([img], diameter=None, channels=channels)
        all_masks.append(masks[0])

    if os.path.exists(output_dir):
        shutil.rmtree(output_dir)
    os.makedirs(output_dir, exist_ok=True)

    for i, file in enumerate(files):
        filename = os.path.basename(file)
        output_path = os.path.join(output_dir, filename)
        imsave(output_path, all_masks[i].astype(np.uint8))

    all_done.append(sub)
    print(f"Finished processing {sub}, masks saved to {output_dir}")

print("\nAll folders processed:", ", ".join(all_done))
```
